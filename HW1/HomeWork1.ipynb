SciPy 버전에 다른 Axis 설정경고를 무시하기 위한 import (코드와 무관합니다.)

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_label=breast_cancer.target
```

코드 가독성증가를 위해 데이터/라벨은 따로 분리한다.

```python
breast_cancer_feature=breast_cancer.feature_names
```

데이터 중 정보들의 이름을 찾아보기위해 feature를 선언한다.

```python
df_data = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer_feature)
df_labels = pd.DataFrame(breast_cancer_label)
```

```python
df_labels
```

만약 라벨을 포함하여 표시하고싶을경우

```python
breast_cancer_df = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer_feature)
breast_cancer_df['label'] = breast_cancer_label
```

```python
from sklearn.model_selection import train_test_split
```

train, test 데이터 분류시 셔플을 위해 사용

```python
X_train,X_test,y_train,y_test = train_test_split(breast_cancer_data, breast_cancer_label, test_size=0.2, random_state=42)
```

train과 test 8:2로 분류작업

```python
print("X_train.shape :", X_train.shape)
print("y_train.shape :", y_train.shape)
print("X_test.shape:", X_test.shape)
print("y_test.shape:", y_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
mymodel = KNeighborsClassifier(n_neighbors=3)
```

KNN 알고리즘 관련 라이브러리 임포트 (K값은 3으로 지정)

```python
mymodel.fit(X_train, y_train)
```

```python
prediction = mymodel.predict(X_test)
mymodel.score(X_test, y_test)
```

결과는 0.9298245614035088점이 나오게 된다.


# K값을 7로 늘리면 다음과 같이 변화한다.

```python
mymodel2 = KNeighborsClassifier(n_neighbors=7)
mymodel2.fit(X_train, y_train)
prediction = mymodel2.predict(X_test)
mymodel2.score(X_test, y_test)
```

결과가 0.956140350877193까지 상승하는것을 알 수 있다

# k값을 3으로 유지한상태로 train test 비율을 7:3으로 할 경우 점수는 아래와 같다.

```python
X_train,X_test,y_train,y_test = train_test_split(breast_cancer_data, breast_cancer_label, test_size=0.3, random_state=42)
mymodel.fit(X_train, y_train)
prediction = mymodel.predict(X_test)
mymodel.score(X_test, y_test)
```

결과가 0.9415204678362573점으로 상승하는것을 확인할 수 있다.

```python
mymodel2.fit(X_train, y_train)
prediction = mymodel2.predict(X_test)
mymodel2.score(X_test, y_test)
```
